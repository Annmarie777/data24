```python
query = """
SELECT AVG(Observation.tip) AS AVG_tip
FROM Observation
;"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT MAX(Observation.total_bill) as Maximum_bill FROM Observation
;"""
pd.read_sql(query, conn)
```

```python
##Count how many people were served on each day.

query = """
SELECT SUM(Observation.size) AS N_of_costumers, Day.day FROM Observation
JOIN Day ON Day.day_id=Observation.day_id
GROUP BY Day.day
;"""
pd.read_sql(query, conn)
```

```python
## Calculate the average total_bill grouped by size.
query = """
SELECT AVG(Observation.total_bill) AS Average_bill, Observation.size AS SIZE FROM Observation
GROUP BY Observation.size
;"""
pd.read_sql(query, conn)

```

```python
## Find the total tip amount for smokers vs. non-smokers.
query = """
SELECT SUM(Observation.tip) AS Total_tip, Smoker.smoker FROM Observation
JOIN Smoker ON Smoker.smoker_id=Observation.smoker_id
GROUP BY Smoker.smoker
;"""
pd.read_sql(query, conn)
```

```python
##Determine the average tip percentage (tip/total_bill) for each day.
query = """
SELECT (Observation.tip)/(Observation.total_bill) AS Average_tip, Day.day AS Day FROM Observation
JOIN Day ON Day.day_id=Observation.day_id
GROUP BY  Day.day
;"""
pd.read_sql(query, conn)
```

# **<font color='red'> For average calculation you should either divide the sum of tip to the sum of total bill or use avg aggregate function on the division of tip to total bill. Minus 1 point </font>**

```python
##List all total_bill values with the corresponding day and time
query = """ SELECT Observation.total_bill AS Total_bill, Day.day AS Day, Time.time AS Time FROM Observation
JOIN Day ON Day.day_id=Observation.day_id
JOIN Time ON Time.time_id=Observation.time_id;"""
pd.read_sql(query, conn)
```

```python
### Find the total number of observations for each gender
query = """
SELECT Sex.sex, COUNT(Observation.sex_id) AS bservations
FROM Observation
JOIN Sex ON Observation.sex_id = Sex.sex_id
GROUP BY Sex.sex;
"""
pd.read_sql(query, conn)
```

```python
##10 Retrieve all tips given during Dinner on Saturday
query10 = """
SELECT Observation.tip, Day.day, Time.time
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
JOIN Time ON Observation.time_id = Time.time_id
WHERE Day.day = 'Sat' AND Time.time = 'Dinner';
"""

pd.read_sql(query10, conn)
```

```python
query = """
SELECT SUM(Observation.tip) as Tip, Time.time FROM Observation
JOIN Time ON Time.time_id=Observation.time_id
JOIN Day ON Day.day_id=Observation.day_id
WHERE Day.day = 'Sat' AND Time.time = 'Dinner'
;"""
pd.read_sql(query, conn)
```

```python
##11
query = """
SELECT MAX(Observation.total_bill) as Largest_bill, Sex.sex FROM Observation
JOIN Sex ON Sex.sex_id=Observation.sex_id
GROUP BY Sex.sex
;"""
pd.read_sql(query, conn)
```
