```python
alltogether = []

page_url = "/page/1/"
while page_url:

    response = requests.get(url + page_url)
    soup = BeautifulSoup(response.content, "html.parser")

    quotes = soup.find_all("div", class_="quote")

    for quote in quotes:
        quote_text = quote.find("span", class_="text").get_text()
        author = quote.find("small", class_="author").get_text()
        tags = [tag.get_text() for tag in quote.find_all("a", class_="tag")]

        alink = quote.find("a")["href"]
        apage = requests.get(url + alink)
        asoup = BeautifulSoup(apage.content, "html.parser")

        born = asoup.find("span", class_="author-born-date").get_text()
        location = asoup.find("span", class_="author-born-location").get_text()
        description = asoup.find("div", class_="author-description").get_text(strip=True)


        alltogether.append({
            "quote": text,
            "author": author,
            "tags": ", ".join(tags),
            "born": born,
            "born_location": location,
            "description": description
        })


    next_page = soup.find("li", class_="next")
    page_url = next_page.find("a")["href"] if next_page else None
```

```python
df = pd.DataFrame(alltogether)
df.to_csv("quotes_scraped.csv", index=False)

print(df.head())
```
